# Human Stereopsis - Binocular Fusion

## Introduction
Human stereopsis is the perception of depth and three-dimensional structure obtained on the basis of visual information deriving from two eyes. Binocular fusion refers to the ability to merge images from both eyes to form a single coherent image.

### Components of Stereopsis
1. **Binocular Disparity**: Differences in the images formed on the left and right retinas.
2. **Depth Perception**: Determining the distance and three-dimensional relationships of objects.
3. **Correspondence Problem**: Matching points in the two retinal images to create depth perception.

## Concept of Binocular Fusion
Binocular fusion is crucial for stereopsis. The brain fuses slightly different images seen by each eye into a single image, which gives depth information. This process involves:
- Matching features in images from both eyes.
- Combining information to compute disparity and estimate depth.

## Simulation in Python
We can simulate stereopsis by generating synthetic images with disparities and calculating depth using triangulation.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create synthetic stereo images
def create_synthetic_images():
    img = np.zeros((200, 300), dtype=np.uint8)
    cv2.circle(img, (150, 100), 30, 255, -1)  # Circle in the center
    img_left = np.roll(img, -10, axis=1)      # Shift left image
    img_right = np.roll(img, 10, axis=1)      # Shift right image
    return img_left, img_right

# Display stereo images
def display_images(img_left, img_right):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Left Image")
    plt.imshow(img_left, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Right Image")
    plt.imshow(img_right, cmap='gray')
    plt.axis('off')

    plt.show()

# Compute disparity map
def compute_disparity(img_left, img_right):
    stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
    disparity = stereo.compute(img_left, img_right)
    return disparity

# Main logic
img_left, img_right = create_synthetic_images()
display_images(img_left, img_right)

disparity_map = compute_disparity(img_left, img_right)

# Display disparity map
plt.figure(figsize=(6, 4))
plt.title("Disparity Map")
plt.imshow(disparity_map, cmap='jet')
plt.colorbar(label="Disparity")
plt.axis('off')
plt.show()
```
